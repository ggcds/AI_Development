#  O que  t-SNE

O **t-Distributed Stochastic Neighbor Embedding (t-SNE)**  um algoritmo poderoso para **reduo de dimensionalidade**, amplamente utilizado para a visualizao de conjuntos de dados de alta dimenso. Desenvolvido por **Laurens van der Maaten** e **Geoffrey Hinton** em 2008, o t-SNE  particularmente bem-sucedido em:

- Capturar a **estrutura local de dados complexos**
- Revelar **agrupamentos em um espao de baixa dimenso** (geralmente 2D ou 3D)

---

##  Como o t-SNE Funciona?

O t-SNE comea com a **compreenso de que em um espao de alta dimenso**, cada ponto de dados pode ser visto como um ponto em um **espao multidimensional**.

O algoritmo transforma as **distncias entre os pontos em probabilidades condicionais**, que representam **similaridades**.

> A ideia  que pontos que so prximos uns dos outros tenham uma **alta probabilidade de serem "vizinhos"**, e aqueles que esto distantes tenham uma **probabilidade baixa**.



##  Por que usar t-SNE?

Utilizamos o **t-SNE** para **simplificar a complexidade dos dados**, permitindo-nos **observar padres e grupos que so difceis** de visualizar em mltiplas dimenses.

Por exemplo, na **anlise de dados genticos**, o t-SNE pode ajudar a **identificar subgrupos de clulas com padres de expresso gnica similares**, apesar da complexidade e da grande dimenso dos dados.

##  Etapas do t-SNE

### 1 Medindo Similaridades

Imagine que cada ponto de dados em um espao de alta dimenso seja uma pessoa em uma festa.  
O **t-SNE** comea calculando o quanto cada pessoa (ponto de dados) prefere ficar perto de cada outra pessoa.  
Essa "preferncia"  medida usando a **distncia entre eles**:  
- Pessoas mais prximas so mais propensas a serem "amigas".
- Isso  convertido em **probabilidades**, com amigos prximos tendo **alta probabilidade** de ficarem juntos, e conhecidos distantes tendo **baixa probabilidade**.

---

### 2 Criando o Mapa

Depois, o **t-SNE** cria um **mapa**, que  uma representao mais simples onde as pessoas (pontos de dados) tambm esto presentes.  
Mas desta vez, o espao  muito menor (como mudar de um grande salo para uma sala pequena).  

 Aqui, ele tenta colocar todos no espao pequeno **mantendo as amizades (similaridades)** tanto quanto possvel.

---

### 3 Comparao de Similaridades

Agora, o **t-SNE** olha para as **probabilidades no espao original** (o grande salo) e no **espao pequeno** (a sala) e verifica quo bem as amizades foram preservadas.

Se duas pessoas que eram prximas no espao original **no estiverem prximas no espao reduzido**, h uma **penalidade**.  
O t-SNE ajusta as posies no espao pequeno tentando **minimizar essas penalidades**.

---

### 4 Otimizao

Este processo de ajustar as posies  como um **jogo de quebra-cabea**, onde se tenta **fazer todos felizes com suas posies relativas**.

O t-SNE faz **muitos pequenos ajustes**, sempre tentando manter amigos prximos, at encontrar uma configurao no espao menor onde as **amizades sejam bem representadas**, de acordo com as **probabilidades originais**.

---

### 5 Resultado Final

Depois de muitos ajustes, o **t-SNE** finaliza com uma configurao onde, em geral, **pessoas que eram prximas no espao de muitas dimenses ainda esto prximas no espao de poucas dimenses**.

 Isso permite **visualizar e entender padres e grupos nos dados** que seriam muito difceis de perceber em um espao de muitas dimenses.



##  Mecnica do t-SNE

###  Probabilidades e Similaridades

O primeiro passo no **t-SNE**  converter as **distncias euclidianas entre os pontos em probabilidades** que somam um para cada ponto.

 Isso  feito usando:
- A **distribuio Gaussiana (normal)** para os dados de entrada de alta dimenso
- A **distribuio t de Student** para o mapa de baixa dimenso

---

###  O Desafio da "Maldio da Dimensionalidade"

A **"maldio da dimensionalidade"** refere-se ao fenmeno onde o aumento das dimenses dos dados leva a problemas como:
- Aumento do espao vazio  
- Distoro das relaes de distncia

O t-SNE aborda esse desafio **focando nas probabilidades de similaridade**, em vez de apenas replicar distncias mtricas.  
 Isso ajuda a **preservar a estrutura local dos dados**.



##  Otimizao por Gradiente Descendente

O **t-SNE** utiliza uma tcnica chamada **gradiente descendente** para **minimizar a diferena entre as probabilidades** no espao de alta dimenso e no espao projetado de baixa dimenso.

 Este processo  conhecido como **minimizao da divergncia de Kullback-Leibler** entre as duas distribuies de probabilidades, que mede **quo bem o espao de baixa dimenso representa o de alta dimenso**.



##  Desafios e Limitaes do t-SNE

###  Sensibilidade aos Hiperparmetros

Um dos desafios ao usar o **t-SNE**  sua **sensibilidade a hiperparmetros** como o *perplexity*, que indica quantos vizinhos prximos cada ponto considera.

 A escolha deste parmetro pode afetar significativamente a **aparncia dos grficos finais**  
 **No existe um valor nico ideal** para todos os conjuntos de dados

###  Limitaes na Interpretao de Clusters

Apesar de o **t-SNE** ser excelente para **visualizar agrupamentos**,  importante **no tirar concluses precipitadas** sobre os "clusters" visualizados.

O t-SNE pode:
- Criar ou **exagerar clusters**
- No representar divises naturais nos dados  
 Isso acontece devido  sua forma de **otimizao** e representao.

###  Custo Computacional e Escalabilidade

O **t-SNE** pode ser **computacionalmente intensivo**, especialmente para conjuntos de dados muito grandes.

Isso ocorre porque o algoritmo **compara todos os pares de pontos** para calcular as probabilidades, o que pode se tornar **impraticvel com o aumento do nmero de pontos**.

##  Aplicaes do t-SNE

###  Biologia Computacional

No campo da **biologia computacional**, o **t-SNE**  frequentemente utilizado para **analisar dados de sequenciamento de clulas nicas**, ajudando os pesquisadores a identificar tipos de clulas com base em seus **perfis de expresso gnica**.

 Este tipo de anlise pode **revelar novos subtipos celulares** e **potenciais alvos teraputicos**.

---

###  Reconhecimento de Imagens

O **t-SNE** tambm  aplicado no **reconhecimento de imagens**, onde pode ser usado para **reduzir a dimenso das representaes de imagem** antes de classificaes ou outras anlises.

 Visualizar essas representaes ajuda a entender **como diferentes categorias de imagens so percebidas por modelos de IA**.



##  t-SNE vs PCA

###  PCA (Anlise de Componentes Principais)

- **Mtodo**: Linear  projeta os dados em direes que maximizam a varincia, sem tentar preservar relaes no-lineares.
- **Vantagens**:
  - Rpido e eficiente para grandes conjuntos de dados
  - Reduz rudos e destaca caractersticas mais importantes dos dados
- **Desvantagens**:
  - No captura complexidades e padres no-lineares
  - Pode perder informaes importantes em dados com estrutura intrincada
- **Melhores usos**:
  - Anlise inicial para obter uma viso geral das principais variaes
  - til em finanas e reas com predominncia de relaes lineares

---

###  t-SNE

- **Mtodo**: No-linear  preserva a estrutura local dos dados ao mapear pontos prximos em alta dimenso para pontos prximos em baixa dimenso.
- **Vantagens**:
  - Excelente para visualizar agrupamentos e padres complexos
  - Preserva a vizinhana local, permitindo visualizaes intuitivas
- **Desvantagens**:
  - Computacionalmente intensivo (especialmente com grandes datasets)
  - Sensvel a hiperparmetros como *perplexidade*
- **Melhores usos**:
  - Anlise exploratria para descobrir padres ocultos
  - Especialmente til em biologia, marketing e campos com relaes no-lineares

##  O hiperparmetro *Perplexity*

Imagine que voc est numa festa e quer decidir com quantas pessoas voc deve conversar.

- Se voc conversar com **muitas pessoas**, ter uma boa ideia geral de quem est na festa.
- Se conversar com **poucas pessoas**, entender bem essas poucas, mas no tanto sobre a festa como um todo.

A **perplexidade** no t-SNE  semelhante a isso:  como se fosse um nmero que voc escolhe para decidir **quantas "pessoas" (ou pontos de dados)** cada ponto deve considerar como seus **vizinhos prximos**.

---

### A *perplexidade* ajuda a determinar:

- **Quantos vizinhos prximos** cada ponto deve considerar:
  -  Um nmero **pequeno**: cada ponto olha s para seus vizinhos muito prximos (conversa com poucos).
  -  Um nmero **grande**: cada ponto considera muitos outros como seus vizinhos (conversa com muitos na festa).

---

###  Como Escolher a *Perplexidade*?

- No muito alta, no muito baixa:
  -  Se for **muito alta**, o t-SNE pode perder detalhes, pois cada ponto tenta ser amigo de quase todos.
  -  Se for **muito baixa**, os pontos podem formar **pequenos grupinhos isolados**, no representando bem os dados.

- **Depende dos dados**: no existe um nmero mgico.  
 Voc precisa testar diferentes valores para ver **qual oferece a melhor viso geral** dos seus dados.
