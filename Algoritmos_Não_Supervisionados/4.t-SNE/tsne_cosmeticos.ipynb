##  Projeto  t-SNE

Uma empresa de cosmticos deseja trabalhar num **algoritmo de recomendao** para seus clientes de e-commerce, de forma a **indicar produtos com base em caractersticas similares**, tais como **tipo de pele, marca e preo**.

Porm, como **a combinao entre marcas, tipo de produtos e demais caractersticas  muito grande**, analisar estes dados antes de propor um algoritmo  extremamente desafiador, e a empresa no gostaria de criar um algoritmo que recomendasse produtos que no faam sentido.

---

Desta forma, iremos desenvolver uma **ferramenta visual** para que esta empresa possa **visualizar seus produtos num chart 2D**, de forma a identificar se produtos prximos em termos de caractersticas esto agrupados.

---

Como comentado anteriormente, este algoritmo possui uma **sensibilidade a hiperparmetros**, especialmente o **perplexity**, e apresentaremos este **chart de forma animada**, com diversos parmetros desse hiperparmetro.

### Treinar o algoritmo t-SNE

```python
# copiar dataframe original
X = df_cosmeticos.copy()
X.drop(columns=['Nome', 'Ingredientes'], axis=1, inplace=True)
```

```python
# Separando variveis numricas e categricas
numeric_features = ['Rating', 'Preco']
categorical_features = ['Marca', 'Tipo']
```

```python
# Definir as transformaes
numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder()
```

```python
# Criar Pre Processador de Transformao
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='passthrough'
)
```

```python
# Transformar os dados
X_transformed = preprocessor.fit_transform(X)
```

```python
# Visualizar Dados
X_transformed
```

```python
# Armazenar resultados do t-SNE em DataFrame
results_df = pd.DataFrame()
```

```python
# Loop de treinamento do algoritmo, mudando o parmetro Perplexity
for perplexity in range(5, 51, 1):

    # Criar e treinar modelo
    tsne = TSNE(n_components=3, perplexity=perplexity, init="random", n_iter=250, random_state=51)
    tsne_results = tsne.fit_transform(X_transformed)

    # Armazenar resultados
    temp_df = pd.DataFrame(tsne_results, columns=['Componente 1', 'Componente 2', 'Componente 3'])
    temp_df['Perplexity'] = perplexity
    results_df = pd.concat([results_df, temp_df], axis=0)
```

```python
results_df.head(10)
```

```python
# Reset no ndice para realizar o plot
results_df.reset_index(drop=True, inplace=True)
```

### Visualizar Resultados

```python
# Criar um ScatterPlot Animado com variao no Perplexity
fig = px.scatter_3d(results_df, x='Componente 1', y='Componente 2', z='Componente 3', animation_frame='Perplexity',
                 title='Visualizao do t-SNE com variao do Perplexity')

fig.show()
```
