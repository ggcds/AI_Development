#  O que  clusterizao hierrquica

A clusterizao hierrquica  uma tcnica distinta dentro dos mtodos de agrupamento em aprendizado de mquina no supervisionado, que se diferencia principalmente:

-  **Pela forma como os clusters so estruturados**
-  **Pela flexibilidade no nmero de clusters**

 especialmente til quando:

-  So feitas **anlises exploratrias**
-  A **relao entre os pontos de dados**  mais relevante do que a definio rgida de grupos
-  Os dados tm **mltiplas escalas ou nveis de agregao**

---

##  No requer a especificao do nmero de clusters

- Enquanto o **K-means** exige definir o nmero de clusters **a priori**...
- ...o agrupamento hierrquico permite que isso seja feito **posteriormente**, com base na:
  -  **Anlise do dendrograma**
  -  **Escolha do ponto de corte**

---

##  Comparao com K-means

###  Vantagens do K-means:
-  **Alta eficincia** com grandes volumes de dados
-  Funciona bem com clusters de **forma esfrica**

###  Limitaes do K-means:
-  **No captura bem clusters de formas complexas ou tamanhos variados**
-  **Solues variam conforme os centroides iniciais**

###  Vantagens do Hierrquico:
-  Identifica **estruturas complexas**
-  Fornece uma **sada mais estvel** com base em hierarquia
-  **Mais suscetvel a outliers e  ordem dos dados**

---



##  Etapas

### 1 Definio de Similaridade
Primeiramente,  necessrio definir uma mtrica de similaridade.  
 Comumente, distncias como a **Euclidiana**, **Manhattan**, ou outras distncias especficas do domnio so usadas para quantificar **quo similares ou distintos so os objetos**.

### 2 Construo da Matriz de Distncia
 Calcula-se a distncia entre cada par de objetos no conjunto de dados, resultando em uma **matriz de distncia**.

### 3 Construo do Dendrograma

-  **Aglomerativo**: Inicialmente, cada objeto  tratado como um cluster individual e os clusters mais prximos so fundidos.
-  **Divisivo**: Inicia-se com um nico cluster que inclui todos os objetos e o dendrograma  construdo **dividindo sucessivamente** os clusters.

### 4 Corte do Dendrograma

 A escolha do nmero de clusters  feita **cortando o dendrograma em uma certa altura**, que define o nmero final de clusters.


#  O que  o algoritmo de clusterizao hierrquica aglomerativo

##  Etapas

### 1 Inicializao
 Cada ponto de dado  tratado como um **cluster individual**.

### 2 Clculo da Matriz de Distncia
 Antes de iniciar o processo de fuso, calcula-se a **matriz de distncia** que contm as distncias entre todos os pares de pontos.

### 3 Fuso de Clusters
 Encontre os **dois clusters mais prximos** um do outro com base na matriz de distncia e **combine em um novo cluster**.

### 4 Atualizao da Matriz de Distncia
 Aps cada fuso,  necessrio atualizar a matriz de distncia para refletir a **distncia entre o novo cluster formado e os demais clusters**.

### 5 Repetio
 O processo de **encontrar os clusters mais prximos**, fundi-los e atualizar a matriz de distncia  repetido at que **todos os pontos de dados estejam no mesmo cluster**.

### 6 Construo do Dendrograma
 Ao longo do processo,  possvel construir um **dendrograma** para:
-  Visualizar a formao de clusters
-  Determinar o **ponto de corte** que define o nmero final de clusters


#  O que  o algoritmo de clusterizao hierrquica divisivo

##  Etapas

### 1 Inicializao
 Comece com um **nico cluster que inclui todos os pontos de dados**.  
Este  o cluster de **nvel mais alto**.

### 2 Clculo da Matriz de Distncia
 Em cada etapa, escolha um cluster para ser dividido.  
A escolha pode ser baseada em critrios como:
-  **Tamanho do cluster**
-  **Heterogeneidade interna**
-  **Mtrica de divisibilidade**

### 3 Identificao do Ponto de Corte
 Determine **como dividir o cluster escolhido em dois subclusters**.

### 4 Execuo da Diviso
 Aplicar o **critrio de diviso** para separar o cluster em dois.  
Pode-se usar algoritmos como **K-Means** ou **DBSCAN** para isso.

### 5 Repetio
 Continue o processo de selecionar e dividir clusters at que:
- Cada ponto seja seu **prprio cluster**, ou
- Se atinja o **nmero desejado de clusters**

### 6 Construo do Dendrograma
 Ao longo do processo,  possvel construir um **dendrograma**, que ilustra como os clusters **so divididos progressivamente**.

#  O que  um dendrograma



##  Aplicaes do Dendrograma

###  Anlise Exploratria
O dendrograma ajuda os analistas a **entenderem a estrutura dos dados**, visualizando como os grupos so formados e **quo semelhantes ou diferentes eles so entre si**.

###  Determinao do Nmero de Clusters
 uma ferramenta crucial para decidir o **nmero adequado de clusters** ao realizar a clusterizao hierrquica, permitindo **ajustar o nvel de granularidade da anlise**.

###  Identificao de Outliers
**Outliers ou pontos anmalos** muitas vezes aparecem como **ramos isolados** no dendrograma, facilitando sua **identificao e anlise**.
