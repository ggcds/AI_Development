#  Deteco de Anomalias

A **deteco de anomalias**  o processo de **identificar padres ou valores** que se  
 **desviam significativamente do comportamento normal** ou esperado em um conjunto de dados.  

Isso pode incluir:
-  **Valores extremos**
-  **Outliers**
-  **Erros de coleta de dados**
-  **Ataques mal-intencionados**

A deteco de anomalias  essencial para:
-  Garantir a **qualidade** e a **confiabilidade** dos dados
-  **Detectar e prevenir ameaas cibernticas**

##  Alguns Casos de Uso

-  **Deteco de Fraude**:  
  Identificar transaes suspeitas em **finanas**, **comrcio eletrnico** e outras indstrias.

-  **Monitoramento e anlise de trfego de rede**:  
  Detectar **ataques cibernticos** como invases, **malware** e **ransomware**.

-  **Controle de qualidade**:  
  Detectar **defeitos** em produtos ou **falhas** em processos industriais.

-  **Segurana**:  
  Identificar comportamentos suspeitos em **sistemas de segurana**, como **acesso no autorizado**.

-  **Doenas e Anomalias Mdicas**:  
  Detectar padres anmalos em **exames de sangue** ou **registros mdicos** que possam indicar uma **condio rara** ou uma **doena**.

-  **Anomalias em Sries Temporais**:  
  Identificar **pontos de dados anmalos** em sries temporais, como dados de **sensores** ou **mtricas de desempenho**.

-  **Anlise de Desempenho de Aplicaes Web**:  
  Identificar **picos de tempo de resposta** ou **falhas no servidor** que no correspondem ao comportamento normal da aplicao.



##  LOF (Local Outlier Factor)

**LOF (Local Outlier Factor)**  uma tcnica de aprendizado de mquina **no supervisionado**  
usada para detectar **anomalias** em conjuntos de dados.

 Ele compara a **densidade de cada ponto** com a densidade dos seus **vizinhos mais prximos**.  
Se um ponto est em uma regio de **densidade significativamente menor** que a de seus vizinhos,  
ele  considerado uma **anomalia**.

 O LOF  til em situaes com **densidades variveis**, permitindo identificar pontos **isolados localmente**,  
mesmo com a presena de **clusters com diferentes densidades**.



##  Como funciona o algoritmo LOF?

###  Passo 1: Definir a quantidade de vizinhos (k)

Defina o nmero de **vizinhos (k)** a serem considerados para cada ponto de dados.

 Determinar este nmero  crucial. Para isso, podemos realizar:
-  Testes empricos com **diversos valores**
-  Anlise de **estabilidade**
-  Visualizao em **grficos de disperso**

Alm disso,  possvel:
-  Comparar a **densidade entre pontos anmalos e normais**  
-  Avaliar as **diferenas locais** entre eles

###  Passo 2: Clculo da Distncia k-Distncia

A **k-distncia**  a **distncia entre um ponto (A) e seu k-simo vizinho mais prximo**.

 Exemplo:  
Se escolhermos `k = 3`, a k-distncia do ponto `A` ser a **distncia entre `A` e seu terceiro vizinho mais prximo**.

 Na figura referida, a **k-distncia de A** seria a **distncia at o ponto E**, que  o **3 vizinho mais prximo**.



###  Passo 3: Clculo do Alcance k-Distncia (Reachability Distance ou RD)

Para dois pontos (A) e (B), o **alcance k-distncia**  definido como:

\[
RD_k(A, B) = \max(\text{kdistncia}(B), \text{distncia}(A, B))
\]

 A distncia de alcance  o **maior valor entre a k-distncia de B e a distncia direta entre A e B**.

 Isso garante que a distncia de alcance:
- Nunca seja menor que a k-distncia de B
-  **Evite distores causadas por valores muito pequenos**
-  Ajude a **normalizar as distncias**
-  **Reduza o impacto de outliers** na anlise de densidade local



###  Passo 4: Clculo da Densidade de Alcance Local (Local Reachability Density ou LRD)

A **densidade de alcance local**  a **inversa da mdia da distncia de alcance k**  
para todos os `k` vizinhos de `A`.  
Ela representa **quo densamente o ponto A est rodeado por seus vizinhos**.

\[
LRD_k(A) = \frac{1}{\frac{1}{k} \sum_{B \in kNN(A)} RD_k(A, B)}
\]

 Onde `kNN(A)`  o conjunto de `k` vizinhos mais prximos de `A`.

###  Passo 5: Clculo do LOF

Compara a **densidade local do ponto A** com a **densidade local de seus vizinhos**.

  calculado como a **mdia das razes** entre:
- As densidades locais dos vizinhos de A (`LRD_k(B)`)
- Pela densidade local de A (`LRD_k(A)`)

\[
LOF_k(A) = \frac{1}{|kNN(A)|} \sum_{B \in kNN(A)} \frac{LRD_k(B)}{LRD_k(A)}
\]

 Onde `kNN(A)` representa o conjunto de **k vizinhos mais prximos** de A.
