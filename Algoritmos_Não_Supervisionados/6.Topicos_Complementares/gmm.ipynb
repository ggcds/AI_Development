##  GMM (Gaussian Mixture Models)

 **Gaussian Mixture Models (GMMs)** so modelos probabilsticos utilizados em problemas de **agrupamento (clusterizao)** de dados, que assumem que os dados so gerados a partir de uma **mistura de vrias distribuies gaussianas**.

 Cada componente da mistura (**cluster**)  uma **distribuio normal (gaussiana)**,  
 e a distribuio total  uma **combinao ponderada dessas distribuies**.



##  GMM (Gaussian Mixture Models) vs K-Means

 GMM possui certa similaridade com o algoritmo **K-Means**, tambm usado para clusterizao.  
 Porm, podemos dizer que o **GMM  uma verso probabilstica do K-Means**.

 Comparado ao GMM, o K-Means possui algumas limitaes como:

-  **Assume que os clusters so esfricos** e de **tamanho similar**, o que nem sempre reflete a realidade.
-   um mtodo de **hard cluster**, ou seja, cada ponto de dado  designado para **apenas um cluster**.



##  Casos de Uso

###  Agrupamento de Dados (Clustering)
GMMs so usados para **agrupar dados em subconjuntos homogneos**, onde cada grupo  modelado por uma **distribuio gaussiana diferente**.

###  Modelagem de Distribuies Complexas
So utilizados para modelar **distribuies de dados multi-modais** ou com **formas complexas** que no podem ser bem representadas por uma nica distribuio gaussiana.A

###  Segmentao de Imagens  
Na anlise de imagens, GMMs so aplicados para **segmentar uma imagem em diferentes regies** com caractersticas semelhantes.

###  Reconhecimento de Padres e Fala  
GMMs so usados em **reconhecimento de fala** para modelar as caractersticas acsticas de diferentes fonemas.

###  Deteco de Anomalias  
Utilizados para **identificar comportamentos anmalos ou raros** em conjuntos de dados, comparando as observaes com a **distribuio modelada**.



##  Vantagens e Desvantagens

###  Vantagens
-  Alta flexibilidade para **modelar dados complexos**.
-  Capacidade de **lidar com dados de diferentes formas e orientaes**.
-  Aplicvel em diversas reas como **viso computacional**, **processamento de fala** e **deteco de anomalias**.

###  Desvantagens
-  Pode **convergir para um timo local** dependendo da **inicializao dos parmetros**.
-  **Sensvel  escolha do nmero de componentes** (K).
-  **Computacionalmente intensivo** para grandes conjuntos de dados.



##  Como funciona o algoritmo GMM?

###  Passo 1: Definir a quantidade de clusters

Para definir a quantidade de clusters, temos um desafio similar ao que ocorre no **K-Means**.  
Desta forma, podemos:

-  Definir este parmetro `k` com base no **conhecimento do domnio** do problema a ser modelado, ou  
-  Utilizar uma **mtrica de validao** para avaliar o modelo.

 No caso dos GMMs, a mtrica mais utilizada  chamada de **BIC (Bayesian Information Criterion)**.

###  Passo 2: Inicializa Mdia, Covarincia e Pesos

-  **Mdia**:  
  Representa o **centro de cada Gaussiana**.  
  Inicializa de forma **aleatria**.

-  **Covarincia**:  
  Define a **forma e orientao** de cada Gaussiana.  
  Tambm  inicializada de forma **aleatria**.

-  **Peso**:  
   a **proporo** que cada componente gaussiano contribui para a mistura total.  
  Inicializa com base no parmetro `k`.  
   *Exemplo*: Se `k = 3`, cada componente  inicializado com `1/3`.

###  Passo 3: Usar o algoritmo EM (Expectation-Maximization)

O algoritmo **EM (Expectation-Maximization)**  uma tcnica usada para encontrar parmetros em modelos estatsticos, como os **Gaussian Mixture Models (GMMs)**.

 Ele  utilizado para **ajustar os parmetros das distribuies gaussianas** (mdias, covarincias e pesos), com o objetivo de **melhor modelar os dados observados**.

 A cada iterao, o modelo se ajusta para **melhorar a representao dos dados**, resultando em uma **melhor compreenso da estrutura subjacente** dos dados.

####  Passo 3a: Etapa Expectation (E)

O passo **E** do algoritmo EM calcula a **responsabilidade** de cada componente da mistura por cada ponto de dado,  
usando as **estimativas atuais dos parmetros** (mdia, covarincia e peso).

 Essas responsabilidades indicam a **probabilidade de que cada ponto pertena a um componente**,  
e sero usadas na prxima etapa do algoritmo (**passo M**) para **atualizar os parmetros do modelo**.

####  Passo 3b: Etapa Maximization (M)

O objetivo do passo **M** do algoritmo EM  **atualizar os parmetros do modelo**  
(mdias , covarincias  e pesos  das distribuies),  
de forma a **maximizar a probabilidade dos dados observados**,  
com base nas **responsabilidades calculadas na Etapa de Expectation**.

###  Passo 4: Itere no algoritmo EM at convergir

 Execute os passos **E** e **M** repetidamente at que ocorra a **convergncia** do algoritmo,  
ou seja, at que os **parmetros do modelo deixem de mudar significativamente** de uma iterao para outra.
