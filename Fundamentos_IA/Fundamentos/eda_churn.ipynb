## ‚ùì Tipos de Dados Ausentes

---

### üü© Dados faltantes completamente ao acaso  
**(MCAR - Missing Completely at Random)**  
> O fato de que um certo valor est√° faltando **n√£o tem nada a ver com seu valor hipot√©tico nem com os valores de outras vari√°veis**.

---

### üü© Dados faltantes ao acaso  
**(MAR - Missing at Random)**  
> A aus√™ncia de dados **n√£o depende dos pr√≥prios valores ausentes**, mas **pode estar relacionada a outras vari√°veis observadas** no dataset.

---

### üü© Dados faltantes n√£o ao acaso  
**(MNAR - Missing Not at Random)**  
> O valor ausente pode **depender do pr√≥prio valor que est√° faltando** (ex: um sal√°rio muito alto n√£o informado),  
> ou **depender do valor de outra vari√°vel** relacionada.

---

> üîç Entender o tipo de dado ausente ajuda a escolher o melhor m√©todo de tratamento: remo√ß√£o, imputa√ß√£o simples, imputa√ß√£o por modelo, etc.

## üîó An√°lise Bivariada

A **an√°lise bivariada** √© uma **t√©cnica estat√≠stica** que se concentra na **rela√ß√£o entre duas vari√°veis** em um conjunto de dados.

---

### üéØ Objetivo

Entender **como uma vari√°vel est√° relacionada √† outra**, geralmente por meio de:

- üìä **Gr√°ficos**
- üìã **Tabelas cruzadas**
- üìà **C√°lculos de correla√ß√£o**

---

### üß† Benef√≠cios

Essa an√°lise permite identificar:

- üîç **Padr√µes**
- ü§ù **Associa√ß√µes**
- üîÑ **Depend√™ncias** entre duas vari√°veis

> Isso fornece **insights** sobre como essas vari√°veis interagem ‚Äî o que √© essencial para a an√°lise de dados e a tomada de decis√µes informadas.


````python
# Hip√≥tese: Clientes com Contrato do Tipo Mensal s√£o mais propensos ao Churn
# construir tabela de contingencia entre tipo de contrato e churn - quantidade
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
````

````python
# construir tabela de contingencia entre tipo de contrato e churn - proporcao
pd.crosstab(df_churn.Churn, df_churn.Contract, normalize='index', margins=True, margins_name='Total')
````

Considerando o publico total, os clientes de contrato mensal representam 55%. Por√©m, se considerarmos apenas os clientes que abandonaram o servi√ßo, 88% tinham contrato mensal. Desta forma, podemos afirmar que h√° uma correla√ß√£o entre o tipo de contrato mensal e o abandono.

## üß™ Interpreta√ß√£o do Teste Qui-Quadrado

- O teste serve para **confirmar ou rejeitar a hip√≥tese nula (H‚ÇÄ)**.

- Se a **probabilidade de observarmos H‚ÇÄ (valor-p)** for **menor que 0.05**, temos evid√™ncias suficientes para **rejeitar a hip√≥tese nula**.

- Neste caso, **aceitamos a hip√≥tese complementar (H‚ÇÅ)**:  
  > As duas vari√°veis **n√£o s√£o independentes**, ou seja, **h√° associa√ß√£o significativa entre elas**.

---

> ‚úÖ **Regra pr√°tica:**  
> Se `p-value < 0.05` ‚Üí rejeitamos H‚ÇÄ  
> Se `p-value ‚â• 0.05` ‚Üí n√£o rejeitamos H‚ÇÄ


````python
# gerar um df da crosstab (sem totais)
df_crosstab_churn__contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
````

````python
# calcular os scores e p-values
from scipy.stats import chi2_contingency
````

````python
# tupla = lista imutavel
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn__contract)
````

````python
chi_scores_churn_contract
````

````python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
````

````python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
````

````python
# apresentar numeros com decimais sem a notacao cientifica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
````

````python
# criando dataframe com os resultados
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-value': pvalues_churn_contract})
````

````python
# analisar scores e p-value
df_chi_scores_churn_contract
````

Conforme Teste Estat√≠stico (Chi-Square), o P-value <= 0.05, desta forma rejeitamos a hip√≥tese nula, ou seja, as vari√°veis n√£o s√£o independentes. Pelo Qui2 alto, podemos afirmar que h√° uma forte correla√ß√£o.

## üß™ Avaliar Correla√ß√£o entre Vari√°vel Qualitativa e Quantitativa

- Utilizar o mesmo **teste Qui-Quadrado (Chi-Square)**  
- Para isso, √© necess√°rio **transformar a vari√°vel quantitativa em qualitativa**

---

### üí° Hip√≥tese:
> **Clientes com menos de 6 meses de contrato s√£o mais propensos ao churn**

---

### üõ†Ô∏è Etapa pr√°tica:
- Criar uma **nova coluna no DataFrame** com base em uma condi√ß√£o l√≥gica sobre outra coluna  
- **Nome da nova coluna**: `'TempoMenor6Meses'`


````python
import numpy as np
````

````python
# criar coluna nova
df_churn['TempoMenor6Meses'] = np.where(df_churn.tenure<6, 'Yes', 'No')
````

````python
# checar o dataframe
df_churn.head(5)
````

````python
# gerar um df da crosstab (sem totais)
df_crosstab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.TempoMenor6Meses)
````

````python
# calcular os scores
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
````

````python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
````

````python
# criando dataframe com os resultados
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'p-value': pvalues_churn_tenure})
````

````python
# analisar os resultados
df_chi_scores_churn_tenure
````

Como o P-value <= 0.05, rejeitamos a H0, ou seja, as vari√°veis n√£o s√£o independentes. E vale mencianos que com base do Qui2 resultante das 2 an√°lises, a correla√ß√£o entre Churn e Tempo de Contrato < 6 meses √© menor do que a correla√ß√£o entre Churn e Tipo de Contrato.

## üö® Outliers (Valores At√≠picos)

Um **outlier** √© um dado que √© muito diferente dos outros dados em um conjunto.  
√â como um **ponto fora da curva**.

## üìê M√©todo de Tukey ‚Äì IQR (Distribui√ß√£o N√£o Normal)

O m√©todo de **Tukey** utiliza o **IQR (Intervalo Interquartil)** para identificar outliers, sendo especialmente √∫til em distribui√ß√µes que **n√£o seguem uma curva normal**.

---

### üî¢ IQR ‚Äì Intervalo Interquartil

√â calculado da seguinte forma:

```python
IQR = Q3 - Q1


````python
q1_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.25)
q3_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.75)
iqr_TotalCharges_month = q3_TotalCharges_month - q1_TotalCharges_month
iqr_TotalCharges_month
````

````python
# limites inferior e superior
limInf_TotalCharges_month = q1_TotalCharges_month - (iqr_TotalCharges_month * 1.5)
limSup_TotalCharges_month = q3_TotalCharges_month + (iqr_TotalCharges_month * 1.5)
limInf_TotalCharges_month, limSup_TotalCharges_month
````

````python
df_churn_month[(df_churn_month.TotalCharges < limInf_TotalCharges_month) | (df_churn_month.TotalCharges > limSup_TotalCharges_month)]
````

## üìä Z-Score (Usado para Distribui√ß√£o Normal)

O **Z-Score** √© uma medida estat√≠stica que indica **quantos desvios padr√£o** um determinado valor est√° distante da m√©dia.

---

### üßÆ F√≥rmula do Z-Score:

```python
zscore = (x - m√©dia) / desvio_padr√£o


````python
from scipy.stats import zscore
````

````python
z = np.abs(zscore(df_churn_month.TotalCharges))
````

````python
z
````

````python
df_churn_month[z > 3.0]
````
