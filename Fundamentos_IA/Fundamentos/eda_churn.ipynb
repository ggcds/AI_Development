## ❓ Tipos de Dados Ausentes

---

### 🟩 Dados faltantes completamente ao acaso  
**(MCAR - Missing Completely at Random)**  
> O fato de que um certo valor está faltando **não tem nada a ver com seu valor hipotético nem com os valores de outras variáveis**.

---

### 🟩 Dados faltantes ao acaso  
**(MAR - Missing at Random)**  
> A ausência de dados **não depende dos próprios valores ausentes**, mas **pode estar relacionada a outras variáveis observadas** no dataset.

---

### 🟩 Dados faltantes não ao acaso  
**(MNAR - Missing Not at Random)**  
> O valor ausente pode **depender do próprio valor que está faltando** (ex: um salário muito alto não informado),  
> ou **depender do valor de outra variável** relacionada.

---

> 🔍 Entender o tipo de dado ausente ajuda a escolher o melhor método de tratamento: remoção, imputação simples, imputação por modelo, etc.

## 🔗 Análise Bivariada

A **análise bivariada** é uma **técnica estatística** que se concentra na **relação entre duas variáveis** em um conjunto de dados.

---

### 🎯 Objetivo

Entender **como uma variável está relacionada à outra**, geralmente por meio de:

- 📊 **Gráficos**
- 📋 **Tabelas cruzadas**
- 📈 **Cálculos de correlação**

---

### 🧠 Benefícios

Essa análise permite identificar:

- 🔍 **Padrões**
- 🤝 **Associações**
- 🔄 **Dependências** entre duas variáveis

> Isso fornece **insights** sobre como essas variáveis interagem — o que é essencial para a análise de dados e a tomada de decisões informadas.


````python
# Hipótese: Clientes com Contrato do Tipo Mensal são mais propensos ao Churn
# construir tabela de contingencia entre tipo de contrato e churn - quantidade
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
````

````python
# construir tabela de contingencia entre tipo de contrato e churn - proporcao
pd.crosstab(df_churn.Churn, df_churn.Contract, normalize='index', margins=True, margins_name='Total')
````

Considerando o publico total, os clientes de contrato mensal representam 55%. Porém, se considerarmos apenas os clientes que abandonaram o serviço, 88% tinham contrato mensal. Desta forma, podemos afirmar que há uma correlação entre o tipo de contrato mensal e o abandono.

## 🧪 Interpretação do Teste Qui-Quadrado

- O teste serve para **confirmar ou rejeitar a hipótese nula (H₀)**.

- Se a **probabilidade de observarmos H₀ (valor-p)** for **menor que 0.05**, temos evidências suficientes para **rejeitar a hipótese nula**.

- Neste caso, **aceitamos a hipótese complementar (H₁)**:  
  > As duas variáveis **não são independentes**, ou seja, **há associação significativa entre elas**.

---

> ✅ **Regra prática:**  
> Se `p-value < 0.05` → rejeitamos H₀  
> Se `p-value ≥ 0.05` → não rejeitamos H₀


````python
# gerar um df da crosstab (sem totais)
df_crosstab_churn__contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
````

````python
# calcular os scores e p-values
from scipy.stats import chi2_contingency
````

````python
# tupla = lista imutavel
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn__contract)
````

````python
chi_scores_churn_contract
````

````python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
````

````python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
````

````python
# apresentar numeros com decimais sem a notacao cientifica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
````

````python
# criando dataframe com os resultados
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-value': pvalues_churn_contract})
````

````python
# analisar scores e p-value
df_chi_scores_churn_contract
````

Conforme Teste Estatístico (Chi-Square), o P-value <= 0.05, desta forma rejeitamos a hipótese nula, ou seja, as variáveis não são independentes. Pelo Qui2 alto, podemos afirmar que há uma forte correlação.

## 🚨 Outliers (Valores Atípicos)

Um **outlier** é um dado que é muito diferente dos outros dados em um conjunto.  
É como um **ponto fora da curva**.

---

### 📌 Exemplo:
Imagine um conjunto de dados com a **altura de 100 pessoas**.  
A média é **1,70m**, mas uma pessoa tem **2,50m**.  
Essa pessoa está muito fora da média e é considerada um **outlier**.

---

### 🧾 Causas Comuns:
- ⚙️ **Erros de medição**
- ❌ **Dados incompletos**
- 🎲 **Eventos aleatórios ou raros**

---

Outliers podem **distorcer análises estatísticas**, por isso é essencial **identificá-los e tratá-los corretamente**.


## 🧹 Lidando com Outliers

---

### 🕵️‍♂️ Identificação e Documentação
> Identifique os outliers nos seus dados e **documente-os**.  
> Compreender a natureza dos outliers é **fundamental** para decisões posteriores.

---

### 🗑️ Remoção
> Em alguns casos, é possível **remover os outliers** do conjunto de dados,  
> desde que isso seja **justificável e não distorça a análise**.

---

### 🔄 Transformação de Dados
> Aplicar **transformações matemáticas** (como logaritmo ou raiz quadrada)  
> pode **reduzir o impacto** dos outliers sem removê-los.

---

### 🧱 Binning (Agrupamento)
> Dividir os dados em **intervalos (bins)** pode ajudar a **reduzir a influência dos outliers**,  
> transformando a variável em **categórica**.

---

### 🔁 Substituição
> Substituir outliers por valores **mais representativos**, como:  
> - Mediana  
> - Média truncada  
> - Valores interpolados

---

### 🤖 Modelagem Robusta
> Utilizar algoritmos de machine learning **robustos a outliers**, como:  
> - Regressão robusta  
> - Árvores de decisão  
> - Métodos ensemble

---

### 🔍 Análise Separada
> Executar análises **com e sem os outliers** para observar o impacto deles nos resultados.

---

### 🧪 Entender a Causa
> Investigar a origem dos outliers para decidir se são:  
> - Erros de medição  
> - Informações legítimas e valiosas

## 📐 Método de Tukey – IQR (Distribuição Não Normal)

O método de **Tukey** utiliza o **IQR (Intervalo Interquartil)** para identificar outliers, sendo especialmente útil em distribuições que **não seguem uma curva normal**.

---

### 🔢 IQR – Intervalo Interquartil

É calculado da seguinte forma:

```python
IQR = Q3 - Q1

## 📊 Z-Score (Usado para Distribuição Normal)

O **Z-Score** é uma medida estatística que indica **quantos desvios padrão** um determinado valor está distante da média.

---

### 🧮 Fórmula do Z-Score:

```python
zscore = (x - média) / desvio_padrão


````python
from scipy.stats import zscore
````

````python
z = np.abs(zscore(df_churn_month.TotalCharges))
````

````python
z
````
