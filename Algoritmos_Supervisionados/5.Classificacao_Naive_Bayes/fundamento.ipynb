## O que  o algoritmo de Naive Bayes?

O algoritmo **Naive Bayes**  um mtodo de aprendizado de mquina que faz previses com base no [Teorema de Bayes](https://pt.wikipedia.org/wiki/Teorema_de_Bayes).

Ele assume que as caractersticas (variveis) usadas para fazer a previso **so independentes entre si**, o que  uma simplificao (da o termo *"naive"* ou *ingnuo*).

O modelo calcula a probabilidade de uma instncia pertencer a uma classe especfica usando a **probabilidade condicional** das caractersticas, dadas essa classe.

---

###  Aplicaes

- Classificao de textos
- Filtros de spam
- Cenrios onde a independncia entre variveis  uma suposio aceitvel

---

###  Vantagens

- **Simples e eficiente**
- **Fcil de implementar**
- Especialmente til **para conjuntos de dados grandes**



Suponha que:

- **A** seja a presena de uma doena  
- **B** seja um resultado positivo em um teste

As probabilidades podem ser interpretadas da seguinte forma:

- **P(A | B)**: Probabilidade de ter a doena **dado** que o teste deu positivo.
- **P(B | A)**: Probabilidade de o teste dar positivo **dado** que a pessoa tem a doena.
- **P(A)**: Probabilidade geral de uma pessoa ter a doena (sem considerar o teste).
- **P(B)**: Probabilidade geral de obter um resultado positivo no teste (incluindo falsos positivos).

---

 Essa interpretao  a base para o uso de **Naive Bayes** em problemas de classificao, como deteco de doenas, filtros de spam, e muito mais!

#  Mtricas de Classificao

##  Acurcia

A **acurcia**  a proporo de instncias **corretamente classificadas** em relao ao **total de instncias**.

> Apesar de fcil interpretao, pode ser **enganosa em conjuntos desbalanceados**, pois favorece a classe majoritria.

**Frmula:**

(TP + TN) / Total de instancias

---

##  Preciso

A **preciso**  a razo de instncias **verdadeiramente positivas** em relao ao **total de instncias classificadas como positivas**.

> A preciso  importante quando o **custo de falsos positivos  alto**.

**Frmula:**

TP / (TP + FP)

---

###  Legenda:
- **TP**: Verdadeiros Positivos (True Positives)  
- **TN**: Verdadeiros Negativos (True Negatives)  
- **FP**: Falsos Positivos (False Positives)  
- **FN**: Falsos Negativos (False Negatives)  

##  Acurcia x Preciso

Um cenrio em que a **preciso**  mais apropriada que a **acurcia** como mtrica de validao ocorre quando h um **desequilbrio significativo nas classes** do conjunto de dados.

 **Exemplo**:  
Suponha que voc esteja desenvolvendo um modelo para detectar fraudes em transaes financeiras, onde:

- A maioria das transaes  **legtima** (classe negativa)  
- Apenas uma pequena porcentagem  **fraudulenta** (classe positiva)

---

###  Problema com a Acurcia

Se o modelo classificar **todas** as transaes como legtimas, a **acurcia ser alta**, pois a maioria das previses estar correta.

No entanto, isso **no  til**, j que o objetivo  **identificar as transaes fraudulentas**.

---

###  Vantagem da Preciso

Neste caso, a **preciso**  isto :

##  Recall

O **Recall** mede a proporo de instncias **verdadeiramente positivas** que foram corretamente identificadas pelo modelo.

>  crucial quando o **custo de falsos negativos  alto** (por exemplo, em diagnsticos mdicos ou deteco de fraudes).

**Frmula:**

TP / (TP + FN)

---

##  F1-Score

O **F1-Score**  a **mdia harmnica entre Preciso e Recall**, fornecendo uma nica mtrica que **equilibra ambas**.

>  especialmente til quando h um **desequilbrio entre as classes**.

**Frmula:**

2 * (Precisao * Recall / (Precisao + Recall))

---

###  Legenda:
- **TP**: Verdadeiros Positivos  
- **FN**: Falsos Negativos  

##  Acurcia x Preciso x Recall

O **Recall**  uma mtrica apropriada quando o custo associado aos **falsos negativos**  alto e o objetivo  **minimizar os casos positivos erroneamente classificados como negativos**.

>  Em cenrios crticos  como deteco de doenas graves ou preveno de fraudes  priorizar o **Recall**  fundamental.

---

###  Exemplo prtico:

Considere um modelo para detectar uma **doena rara**:

- Se o modelo classificar erroneamente um paciente **doente como saudvel** (falso negativo), as consequncias podem ser **graves**.
- Nesse caso, **maximizar o Recall**  essencial para garantir que a maioria dos casos positivos seja identificada corretamente, mesmo que isso gere alguns falsos positivos.

---

###  Concluso:

-  **Escolha Recall**: Quando o foco  **identificar a maioria dos casos positivos**, especialmente quando **falsos negativos tm consequncias srias**.
-  **Acurcia**: Pode ser **enganosa em conjuntos desbalanceados**.
-  **Preciso**:  melhor quando o **custo dos falsos positivos**  elevado.

## Acurcia x Preciso x Recall x F1-Score

O **F1-Score**  uma mtrica apropriada quando h um **equilbrio desejado entre Preciso e Recall**, e **ambas so igualmente importantes**.

 **Exemplo:**  
Classificao de e-mails como **spam** ou **no-spam**.

---

###  Cenrio:

 crucial:

- Identificar a **maioria dos e-mails de spam**  **Recall**
- Garantir que os e-mails classificados como spam sejam **realmente spam**  **Preciso**

A **acurcia pode ser enganosa** nesse contexto, principalmente se a maioria dos e-mails **no for spam**.

 O **F1-Score** considera **tanto a Preciso quanto o Recall**, gerando uma **mtrica nica e equilibrada**.

---

###  Concluso:

- Use o **F1-Score** quando for necessrio **equilibrar Preciso e Recall**.
- Ele  ideal para cenrios onde **ambas as mtricas so crticas** para avaliar a performance do modelo.
